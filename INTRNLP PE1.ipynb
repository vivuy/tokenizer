```python
import re

from collections import Counter
```

```python
lines = [line.rstrip('\n') for line in open('S17 Text File.txt')]
```

```python
lines.pop(0)
```

```python
items = []

for line in lines:
    result = re.findall(r'([\w+’()-]*)', line)
    for i in result:
        if len(i.strip('’')) > 0:
            items.append(i.strip())
```

```python
#remove parentheses

items = [re.sub(r'[()]', '', x) if '(' in x else x for x in items]
```

```python
#remove quotations in quoted words

items = [re.sub(r'’', '', x) if re.fullmatch(r'(.*?)’', x) != None else x for x in items]
```

```python
#change to lowercase

items = [x.lower() for x in items]
```

```python
#count words

word_count_dict = Counter(items)
```
